```python
# 2a. Exploratory Analysis
customers_df.info(), # customers_df['customer_id'].describe() - Using as another guide
customer_id = "customer_id"
print('\n' f"Descriptive statistics for the '{customer_id}' column:") # adding a carriage return for better visibility
print(customers_df[customer_id].describe())

# 2b Obtain summary statistics for numerical columns
# customers_df["customer_id"].dtypes - Using this to test different formatting
```

```python
inventory_df.info(), # inventory_df['inventory_id'].describe()

inventory_id_col = "inventory_id"
store_id_col = "store_id"
product_id_col = "product_id"
quantity_in_stock_col = "quantity_in_stock"
reorder_level_col = "reorder_level"

# Print descriptive statistics for the specified columns
print('\n' f"Descriptive statistics for the numerical columns in my inventory table:") 
print(inventory_df[[inventory_id_col, store_id_col, product_id_col, quantity_in_stock_col, reorder_level_col]].describe()),
inventory_df.head(1) # using this as a guide
```

```python
products_df.info(), # products_df['product_id'].describe()

product_id_col = "product_id"
price_col = "price"
cost_col = "cost"
weight_col = "weight"


# Print descriptive statistics for the specified columns
print('\n' f"Descriptive statistics for the numerical columns in my products table:") 
print(products_df[[product_id_col, price_col, cost_col, weight_col]].describe()),
products_df.head(1) # using this as a guide
```

```python
stores_df.info(), # stores_df['store_id'].describe() - using this top gather more info on the DF

store_id_col = "store_id"
size_sqft_col = "size_sqft"

# Print descriptive statistics for the specified columns
print('\n' f"Descriptive statistics for the numerical columns in my stores table:") 
print(stores_df[[store_id_col, size_sqft_col]].describe()),
stores_df.head(1) # using this as a guide
```

```python
sales_df.info(), # sales_df.describe() - using as a guide

sale_id_col = "sale_id"
store_id_col = "store_id"
customer_id_col = "customer_id"
product_id_col = "product_id"
quantity_col = "quantity"
total_col = "total"

# Print descriptive statistics for the specified columns
print('\n' f"Descriptive statistics for the numerical columns in my sales table:") 
print(sales_df[[sale_id_col, store_id_col, customer_id_col, product_id_col, quantity_col, total_col]].describe()),
sales_df.head(1) # using this as a guide
```
